```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Tim Cook is the CEO of Apple."
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from transformers import pipeline

summarizer = pipeline(
    task="summarization",
    model="facebook/bart-large-cnn"
)

text = """
Pandas is an open-source Python library that is widely used for data manipulation and analysis.
One of the most popular features of Pandas is the DataFrame.
It is a two-dimensional table-like data structure that allows you to store and manipulate data in a way that is similar to a spreadsheet.
In this tutorial, we will be learning how to retrieve the stock data in pandas dataframe format, dataframe structure, selecting columns, selecting rows, filtering data, adding a column, removing a column, retrieving current trading price & prev day close price, and visualizing data using a simple line chart."""

summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

print(summary[0]['summary_text'])
```

```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
)

texts = [
    "I love this phone!",
    "This phone is sick 🔥",
    "Nice service 🙄",
    "Battery died in 2 hours. Great.",
    "Worst experience ever."
]

for text in texts:
    result = classifier(text)
    print("Text:", text)
    print("Output:", result)
    print("-" * 40)
```
